Segmenting and Clustering Neighborhoods in Toronto 

```python
import pandas as pd
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
df
```

```python
df1=df[df.Borough!='Not assigned']
df1
```

```python
df2=df1.groupby(['Postal code','Borough'])['Neighborhood'].apply(','.join).reset_index()
df2
```

```python
df2.shape
```

```python
df3= pd.read_csv("http://cocl.us/Geospatial_data")
df3
```

```python
df3.rename(columns={'Postal Code':'Postal code'},inplace=True)
df3


```

```python
merged = pd.merge(df2, df3, on='Postal code')
merged
```
